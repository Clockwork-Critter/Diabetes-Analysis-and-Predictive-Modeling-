```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("../data/processed/diabetes_binary_health_indicators_BRFSS2015.csv")
sns.set_theme()
boxplot = sns.boxplot(df, y= df.BMI, x=df.Diabetes_binary)
boxplot.set_title('BMI by Presence of Diabeties')
newplot = sns.histplot(df.Diabetes_binary)
newplot.set_title("Count of Diabeties Status")
newplot.set_ylabel('Count')
g = sns.FacetGrid(df)
sns.histplot(df.HighBP).set_title('High BP')
```
